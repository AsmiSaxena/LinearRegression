```python
import pandas as pd
```

```python
df = pd.read_csv('/content/Salary_Data.csv')
```

```python
df
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

# Create some data
salary = df.Salary
years_experience = df.YearsExperience

# Plot the data
plt.scatter(years_experience, salary)

# Set the title and axis labels
plt.title('Scatter Plot of Salary vs Years Experience')
plt.xlabel('Years Experience')
plt.ylabel('Salary')

# Show the plot
plt.show()
```

```python
x = df.YearsExperience.values.reshape(-1, 1)
y = df.Salary
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
regr.fit(x,y)
```

```python
print('m', regr.coef_[0])
print('b', regr.intercept_)
```

```python
y_predict = regr.predict(x)
plt.scatter(x, y)
plt.plot(x, y_predict, color='red')

plt.show()
```

```python
import numpy as np
```

```python
X_future = np.array(range(15,30 )).reshape(-1, 1)
future_predict = regr.predict(X_future)
plt.scatter(x,y)
plt.plot(X_future, future_predict, color='red')
plt.show()
```
